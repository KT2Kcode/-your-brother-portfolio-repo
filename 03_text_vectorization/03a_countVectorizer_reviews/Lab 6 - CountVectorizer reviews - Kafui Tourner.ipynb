#### print the One Hot vector for the word: "necessary"

```python
import numpy as np  # Add this import at the beginning of your code

word = "necessary"  # Replace with your desired word

if word in vectorizer.vocabulary_:
    index = vectorizer.vocabulary_[word]
    
    # Create the one-hot vector
    one_hot = np.zeros(len(vectorizer.vocabulary_))
    one_hot[index] = 1
    print("One-hot vector for 'necessary':", one_hot)
else:
    print(f"The word '{word}' is not in the vocabulary")
```

### Vectorize text data. Use the variables: countvect, x_counts, x_train_df, x_test_df

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# 1. Split into train/test
x_train, x_test, y_train, y_test = train_test_split(
    df['sentence'], df['label'], test_size=0.2, random_state=42
)

# 2. Initialize CountVectorizer
countvect = CountVectorizer()

# 3. Fit the vectorizer on training data
x_train_counts = countvect.fit_transform(x_train)

# 4. Transform the test data
x_test_counts = countvect.transform(x_test)

# 5. Create DataFrames
x_train_df = pd.DataFrame(x_train_counts.toarray(), columns=countvect.get_feature_names_out())
x_test_df  = pd.DataFrame(x_test_counts.toarray(),  columns=countvect.get_feature_names_out())

# Optional: vectorized full dataset
x_counts = countvect.fit_transform(df['sentence'])
```

#### Training the model

### use the Naive Bayes classifier. Use a variable called "clf" and fit the data (x_train_df,train_set.label)

```python
from sklearn.naive_bayes import MultinomialNB

# Step 1: Initialize the classifier
clf = MultinomialNB()

# Step 2: Fit the model to your training data
clf.fit(x_train_counts, y_train)
```

### Test the model and print the accuracy of the Naive Bayes Classifier

```python
from sklearn.metrics import accuracy_score

# Predict on the test set
predictions = clf.predict(x_test_counts)   # use the same representation as training

# Calculate accuracy
accuracy = accuracy_score(y_test, predictions)

print("Naive Bayes Classifier Accuracy:", accuracy)
```
